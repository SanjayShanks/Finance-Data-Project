# About The Data :
* OPEN  : the price of the stock when market opens on that particular day
* CLOSE : the price of the stock when market opens on that particular day
* HIGH  : the maximum price of the stock on that particular day
* LOW   : the maximum price of the stock on that particular day
* VOLUME: the Volume of stocks trad eon that particular day
* ADJ CLOSE : It is the price at which the last lot of the stock is bought or sold in the last trading session.The adjusted closing price takes into account all the factors that may have affected the stock price after the market hours.
* RETURNS   : (CLOSE / OPEN) - 1



```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import datetime
%matplotlib inline
import yfinance as yf
```

---------------------------------------------------------DATA-------------------------------------------------------------------

```python
start = datetime.datetime(2012, 1, 1)
end = datetime.datetime(2022, 1, 1)
```

```python
yf.pdr_override()
#APPLE
AAPL = data.get_data_yahoo('AAPL', start, end)
#TESLA
TSLA = data.get_data_yahoo('TSLA', start, end)
#AMAZON
AMZN = data.get_data_yahoo('AMZN', start, end)
#MICROSOFT
MSFT = data.get_data_yahoo('MSFT', start, end)
#GOOGLE
GOOG = data.get_data_yahoo('GOOG', start, end)
#NETFLIX
NFLX = data.get_data_yahoo('NFLX', start, end)

#or

#df = data.get_data_yahoo(['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC'],start, end)
```

```python
companies=['AAPL', 'TSLA', 'AMZN', 'MSFT', 'GOOG', 'NFLX']
```

```python
stocks=pd.concat([AAPL, TSLA, AMZN, MSFT, GOOG, NFLX],axis=1,keys=companies)
#axis=1 : to concat on columns
```

```python
stocks
```

```python
#Column names
stocks.columns.names = ['Companies','Stock Info']
```

```python
#checking the heads:
stocks.head()
```

```python
# MAXIMUM CLOSE PRICE OF EACH COMPANY
stocks.xs(key='Close',axis=1,level='Stock Info').max()
#.xs : Multi Indexing concept 
```

```python
# Creating a seperate dataframe to calculate RETURNS
returns = pd.DataFrame()
```

```python
#calculating returns
for company in companies:
    returns[company+' Return'] = stocks[company]['Close'].pct_change()
returns.head()

#pct_change():returns a DataFrame with the percentage difference between the values for each row and,
             #by default, the previous row.
#pct_change(): simply caluclates returns
#returns = (prev value/ current value) - 1
```

-------------------------------------------------------------------------------------------
MOVING AVERAGE(MA):A moving average (MA) chart is a tool used by technical analysts to track the price movements of a security. It plots average prices over a defined period of time.

Plot the rolling 30 day average against the Close Price for AMAZON's stock for the year 2008

```python
plt.figure()
AMZN['Close'].loc['2019-01-01':'2020-01-01'].rolling(window=30).mean().plot(label='30 Day Avg',figsize=(12,6))
AMZN['Close'].loc['2019-01-01':'2020-01-01'].plot(figsize=(12,6))
plt.legend()
```
